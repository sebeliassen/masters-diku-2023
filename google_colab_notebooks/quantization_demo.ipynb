```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 3, 128)
data[0] = 0
data[1] = 3
```

```python
from matplotlib.markers import MarkerStyle
hist_data, bin_edges = np.histogram(data, bins=21)

positions = [0, 1, 2, 3]
labels = ['0b00', '0b01', '0b10', '0b11']

color_percentiles = [0, 0.25, 0.5, 0.75]
colors = [plt.get_cmap('hsv')(p) for p in color_percentiles]


# Calculate bar widths
bar_width = (bin_edges[1] - bin_edges[0])

## Plot the bars with colors from the 'hsv' colormap
n_bins = len(hist_data)
for i, (bar_height, bar_left_edge) in enumerate(zip(hist_data, bin_edges[:-1])):
    bar_color = plt.get_cmap('hsv')(i / (n_bins - 1) * 0.75)
    plt.bar(bar_left_edge, bar_height, width=bar_width, color=bar_color, align='edge')

# Add vertical lines consisting of scattered points with colors and a slight black border

n_points = 20
y_min, y_max = plt.gca().get_ylim()
y_values = np.linspace(y_min, y_max, n_points)
for pos, label, color in zip(positions, labels, colors):
    plt.scatter([pos] * n_points, y_values, s=100, color='black', edgecolor='black', linewidth=3.5, marker="|") # Black border
    plt.scatter([pos] * n_points, y_values, s=45, color=color, edgecolor='black', linewidth=2, label=label, marker="|", edgecolors='black') # Colored points


# Add the legend
plt.title("$Quant(h_v^l)$ visualized")
plt.legend()


plt.show()
```
